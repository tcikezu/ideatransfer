```python
from community import community
from transfer import transfer
import mdp, util, utilRL, copy, time, random, pickle
import numpy as np
from plotCommunity import plotCommunity
```

```python
# Return i in [0, ..., len(probs)-1] with probability probs[i].
def sample(probs):
    target = random.random()
    accum = 0
    for i, prob in enumerate(probs):
        accum += prob
        if accum >= target: return i
    raise Exception("Invalid probs: %s" % probs)
```

```python
def simulate(rl, numTrials=1, maxIterations=100, verbose=False,sort=False):
    T = maxIterations
    
    totalRewards = []  # The rewards we get on each trial
    dataX = np.zeros((T,rl.c.numberMembers))
    dataY = np.zeros((T,rl.c.numberMembers))
    dataZ = np.zeros((T,rl.c.numberMembers))
    posX = np.zeros((T,rl.c.numberMembers))
    posY = np.zeros((T,rl.c.numberMembers))
    ideaDist = np.zeros((T,rl.c.domainSize*2+1))
    
    for trial in range(numTrials):
        # Form a brand new community
        c = community(rl.numberMembers, rl.numberIdeas, rl.domainSize)

        # Form a brand new agent  
        agentMDP = mdp.agentMDP(c)
        
        # Update QLearning community to new community
        rl.c = agentMDP.c
        rl.index = agentMDP.index

        state = agentMDP.startState()
        sequence = [state]
        totalDiscount = 1
        totalReward = 0
        for t in range(maxIterations):
            action = rl.getAction(state)
            transitions = agentMDP.succAndProbReward(state, action)
#             if sort: transitions = sorted(transitions)
#             if len(transitions) == 0: # end state reached
#                 rl.incorporateFeedback(state, action, 0, None)
#                 break

            # Choose a random transition
#             if action[0] == 'move':
#                 newPosition = action[1]
#                 c.updatePosition(agentMDP.index, np.array(newPosition) + c.allPositions[agentMDP.index])
#                 agentMDP.stepForward(c)

#             if action[0] == 'interact':
#                 i = c.getMemberFromAgreement(agentMDP.sampleMembers, agentMDP.index, action[1])
#                 c.ideaDistribution[i] = (1 - c.allGamma[i]) * c.ideaDistribution[i] + c.allGamma[i]*c.ideaDistribution[agentMDP.index]
#                 c.ideaDistribution = util.normalizeDistribution(c.ideaDistribution)
#                 agentMDP.stepForward(c)

#                     dataX[t+n,:] = c.allIdeas[:,0]
#                     dataY[t+n,:] = c.allIdeas[:,1]
#                     dataZ[t+n,:] = c.allIdeas[:,2]

#                     posX[t+n,:] = c.allPositions[:,0]
#                     posY[t+n,:] = c.allPositions[:,1]

#                     ideaDist[t+n,:] = c.ideaDistribution[0][0]   
#             agentMDP.c = c
            if trial == numTrials - 1:
                dataX[t,:] = c.allIdeas[:,0]
                dataY[t,:] = c.allIdeas[:,1]
                dataZ[t,:] = c.allIdeas[:,2]
                posX[t,:] = c.allPositions[:,0]
                posY[t,:] = c.allPositions[:,1]
                ideaDist[t,:] = c.ideaDistribution[0][0]
            
            i = sample([prob for newState, prob, reward in transitions])
            newState, prob, reward = transitions[i]

#             newState = agentMDP.observeCommunity()

#             sequence.append(action)
#             sequence.append(reward)
#             sequence.append(newState)

#             rl.incorporateFeedback(state, action, reward, newState)
            totalReward += totalDiscount * reward
            totalDiscount *= agentMDP.discount()
            state = newState
        
        if verbose:
            print(("Trial %d (totalReward = %s): %s" % (trial, totalReward, sequence)))
        totalRewards.append(totalReward)
    data = [dataX, dataY, dataZ, posX, posY, ideaDist, totalRewards]
    return data
```

```python
c = community(100,4,10)
agentMDP = mdp.agentMDP(community = c, index=0, sampleSize=10)
agentMDP.T = 100
```

```python
rl = mdp.QLearningAlgorithm(agentMDP.index, agentMDP.c, agentMDP.actions, agentMDP.discount(), mdp.communityFeatureExtractor, explorationProb = 0.2)            
```

```python
data = simulate(rl, numTrials=1, maxIterations=100)
```
