```python
from community import community
from plotCommunity import plotCommunity
import pickle # or cloudpickle
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
# with open('data_radius_0.4_N_400.pkl','rb') as infile:
#     data = pickle.load(infile)

from simulateWithoutAgent import simulate
c = community(100,4,10)
# c.allRadii = (np.random.normal(0.2,0.1,c.numberMembers)**2 + np.random.normal(0.2,0.1, c.numberMembers)**2)**0.5
data = simulate(c,T=200)
[dataX, dataY, dataZ, posX, posY, ideaDist] = data
```

```python
fig = plt.figure(figsize=(10,5))
# ax1 = fig.add_subplot(111,projection='3d')
# ax1.scatter3D(dataX[-1], dataY[-1], dataZ[-1])

# ax1Ticks = np.linspace(-c.domainSize,c.domainSize,5)
# ax1bounds = (-c.domainSize,c.domainSize)

i = 199
ax2 = fig.add_subplot(121)
ax2.scatter(posX[i],posY[i],s=1000*c.allRadii**2,alpha=0.08)
ax2bounds = (-2*c.getPositionBounds(),2*c.getPositionBounds())
ax2Ticks = np.linspace(-c.getPositionBounds(),c.getPositionBounds(),4)*1 
ax2.set_xticks(ax2Ticks)
ax2.set_yticks(ax2Ticks)
ax2.set(xlim = ax2bounds, ylim = ax2bounds)
ax2.scatter(posX[i][0], posY[i][0],s=1000*c.allRadii[0]**2,c='r',alpha=0.2)

ax3 = fig.add_subplot(122)
ax3.scatter(c.domain, ideaDist[i])

plt.draw()
plt.savefig('dists'+str(i),dpi=300)
plt.show()
```
